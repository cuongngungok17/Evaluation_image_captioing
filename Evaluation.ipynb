#**Calculate Bleu, Cider, Rouge, Meteor**

```python
from pycocotools.coco import COCO
import sys
sys.path.append('/content/drive/MyDrive/image_captioning/pycocoevalcap')
from eval import COCOEvalCap

annotation_file = '/content/drive/MyDrive/image_captioning/captions_val2017.json'
results_file = '/content/drive/MyDrive/image_captioning/cnn_lstm_coco_cap_4.json'

# create coco object and coco_result object
coco = COCO(annotation_file)
coco_result = coco.loadRes(results_file)

# create coco_eval object by taking coco and coco_result
coco_eval = COCOEvalCap(coco, coco_result)

# evaluate on a subset of images by setting
# coco_eval.params['image_id'] = coco_result.getImgIds()
# please remove this line when evaluating the full validation set
coco_eval.params['image_id'] = coco_result.getImgIds()

# evaluate results
# SPICE will take a few minutes the first time, but speeds up due to caching
coco_eval.evaluate()

# print output evaluation scores
for metric, score in coco_eval.eval.items():
    print(f'{metric}: {score:.3f}')
```

#**Calculate Self Cider**

```python
import json
import os
# from pycocoevalcap.eval import COCOEvalCap as COCOEval
# from pycocoevalcap.PSA_STS import COCOEvalCap as COCOEval
# from pycocoevalcap.diversity_eval import LSA as COCOEval

# load the configuration file
config = json.loads(open('/content/drive/MyDrive/image_captioning/DiversityMetrics/params.json', 'r').read())

pathToData = config['pathToData']
refName = config['refName']
candName = config['candName']
# resultFile = config['resultFile']
resultFile = 'selfcider'
import sys

if 'LSA' in resultFile:
    print('Computing LSA...')
    sys.path.append('/content/drive/MyDrive/image_captioning/DiversityMetrics/pycocoevalcap')
    from diversity_eval import LSA as COCOEval
elif 'selfcider' in resultFile:
    print('Computing selfcider...')
    sys.path.append('/content/drive/MyDrive/image_captioning/DiversityMetrics/pycocoevalcap')
    from diversity_eval import SelfCider as COCOEval
else:
    raise IOError
dfMode = config['idf']
num = int(config['num'])

# Print the parameters
print ("Running metrics with the following settings")
print ("*****************************")
print ("Reference File:%s" % (refName))
print ("Candidate File:%s" % (candName))
print ("Result File:%s" % (resultFile))
print ("IDF:%s" % (dfMode))
print ("num:%d" % (num))
print ("*****************************")
num = 5
# resultFile = resultFile.split('.')[0]+str(num) + '.json'

# calculate metric scores
scorer = COCOEval(pathToData, candName, num=num, dfMode = dfMode)
# scores: dict of list with key = metric and value = score given to each
# candidate
scorer.evaluate()
scores = scorer.eval

# print scores

# write results to output file
with open(os.path.join(pathToData, resultFile), 'w') as outfile:
    json.dump(scores, outfile)
```
